```python
numbers = [3, 0, 7, 9, -3, 9, 9]
def find_adjacent_positive_pairs(arr):
  first_pair = None
  last_pair = None
  for i in range(len(arr) - 1):
    if arr[i] > 0 and arr[i + 1] > 0:
      if first_pair == None:
        first_pair = (i, i + 1)
      last_pair = (i, i + 1)
  return first_pair, last_pair

first, last = find_adjacent_positive_pairs(numbers)
if first:
  print(f"индексы первой пары равны {first[0]} и {first[1]}")
  print(f"индексы последней пары равны {last[0]} и {last[1]}")
else:
  print("в массиве нет соседних положительных чисел")
```

```python
def cyclic_shift_left(arr):
  if len(arr) <= 1:
    return arr
  temp = arr[0]
  for i in range(len(arr) - 1):
    arr[i] = arr[i + 1]
  arr[-1] = temp
  return arr



N = int(input("size = "))
A = []
for num in range(N):
  A.append(int(input("enter number = ")))

A = cyclic_shift_left(A)
print(A)
```

```python
def count_numbers_multiple_seven(arr, n, m):
  count = 0
  for i in range(n):
    for j in range(m):
      if(arr[i][j] % 7 == 0):
        count += 1
  return count

n = int(input("n = "))
m = int(input("m = "))
A = []
for i in range(n):
  row = []
  for j in range(m):
    row.append(int(input(f"enter number with index [{i},{j}] = ")))
  A.append(row)

count = count_numbers_multiple_seven(A, n, m)
print(count)
```

```python
from collections import defaultdict
import re
def word_frequency(str):
  words = re.findall(r'\w+', str.lower())
  frequency = defaultdict(int)

  for word in words:
    frequency[word] += 1
  return dict(frequency)

str = "Hello, i have task. this TASK is very hard"
result = word_frequency(str)
print(result)
```

```python
from collections import defaultdict
def invert_dict(curr_dict):
  invert_curr_dict = defaultdict(list)
  for key, value in curr_dict.items():
    invert_curr_dict[value].append(key)
  return dict(invert_curr_dict)
mydict = {
    "red" : 4,
    "yellow" : 3,
    "blue" : 4,
    "black" : 5
}

new_dict = invert_dict(mydict)
print(new_dict)
```

```python
def merge_dictionaries(first_dict, second_dict):
  new_dict = first_dict.copy()
  for key, value in second_dict.items():
    if key in new_dict:
      new_dict[key] += value
    else:
      new_dict[key] = value
  return new_dict

first_dict = {
    "red" : 4,
    "yellow" : 3,
    "blue" : 4,
    "black" : 5
}
second_dict = {
    "grey" : 7,
    "yellow" : 6,
    "green" : 2,
    "black" : 1
}

result = merge_dictionaries(first_dict, second_dict)
print(result)
```

```python
def find_max_in_dict(curr_dict : dict[any, int]) -> list[any] | None:
  if not curr_dict: return None
  keys = []
  max_value = None
  for key, value in curr_dict.items():
    if max_value == None or value > max_value:
      max_value = value
      keys = [key]
    elif value == max_value:
      keys.append(key)

  return keys

my_dict = {
        "red" : 1,
    "yellow" : 10,
    "blue" : 10,
    "black" : 6
}
print(find_max_in_dict(my_dict))
```

```python
from collections import defaultdict
def group_keys_by_length(dictionary):
  new_dict = defaultdict(list)
  for key in dictionary.keys():
    new_dict[len(key)].append(key)
  return dict(new_dict)

my_dict = {
        "red" : 1,
    "yellow" : 10,
    "blue" : 10,
    "black" : 6,
    "green" : 3
}

result = group_keys_by_length(my_dict)
print(result)
```

```python
from collections import defaultdict
def union_dict_by_value(first_dict, second_dict):
  new_dict = defaultdict(list)
  for d in (first_dict, second_dict):
     for key, value in d.items():
       new_dict[key].append(value)
  return dict(new_dict)

first_dict = {
        "red" : 1,
    "yellow" : 10,
    "blue" : 10,
    "black" : 6,
    "green" : 3
}

second_dict = {
        "sky" : 4,
    "yellow" : 9,
    "blue" : 1,
    "orange" : 6,
    "green" : 2
}

result = union_dict_by_value(first_dict, second_dict)
print(result)
```

```python
def find_unique_values(dictionary):#работает с нехэшируемыми типами
    unique = []
    for value in dictionary.values():
        if value not in unique:
            unique.append(value)
    return unique

my_dict = {
        "sky" : 4,
    "yellow" : 9,
    "blue" : 1,
    "orange" : 4,
    "green" : 9
}

result = find_unique_values(my_dict)
print(result)
```

```python
def max_in_set(current_set):
  if not current_set:
    return None
  maximum = next(iter(current_set))
  for num in current_set:
    if num > maximum:
      maximum = num
  return maximum
my_list = [1, 3, 5, 3, 10, 1, 5, 6]
print(max_in_set(my_list))
```

```python
def split_tuple(my_tuple):
  if not my_tuple: return (), ()
  tup_len = len(my_tuple)
  mid = tup_len // 2
  return my_tuple[:mid], (my_tuple)[mid:]

my_tuple = (3, 5, 2, 6, 1, 8)
fir_tup, sec_tup = split_tuple(my_tuple)
print(fir_tup)
print(sec_tup)
```
